```python
import folium
import pandas as pd
```

```python
clean_data = pd.read_excel('clean_data.xlsx',index_col=0)
```

```python
coord_data = pd.read_excel('coord_data.xlsx',index_col=0)
```

```python
clean_data['lat']=list(coord_data['lat'])
clean_data['long']=list(coord_data['long'])
```

```python
index = [i for i in range(len(clean_data))]
clean_data['Index'] = index
clean_data.set_index('Index',inplace=True)
```

```python
from folium.plugins import FastMarkerCluster

# Create a map object
map_object = folium.Map(location=[55.67543919215466, 12.569927949838403], zoom_start=12)

# Create a FastMarkerCluster object
marker_cluster = FastMarkerCluster([], name='Marker Cluster')

# Iterate through your data points and add them to the marker cluster
for i in range(len(clean_data)):
    folium.Marker(
        location=[clean_data['lat'][i], clean_data['long'][i]],
        popup=clean_data['price'][i]
    ).add_to(marker_cluster)

# Add the marker cluster to the map
marker_cluster.add_to(map_object)

# Display the map
display(map_object)
```
